```
model_results = pd.DataFrame(columns = ['Model_Name','Accuracy', 'F1-Score'])
```

#### ___Performance/Evaluation Metrics___

```
# Defining the metrics
def metrics(model_name, y_test, pred):
    
    print('Accuracy: ', round(accuracy_score(y_test, pred)*100,2))
    
    cm = confusion_matrix(y_test, pred)
    plt.title('Heatmap of Confusion Matrix', fontsize = 15)
    sns.heatmap(cm, annot = True)
    plt.show()
    
    global model_results
    model_results = model_results.append({'Model_Name': model_name , 
                                          'Accuracy': round(accuracy_score(y_test, pred)*100,2), 
                                          'F1-Score':round(f1_score(y_test, pred)*100,2)}, 
                                         ignore_index = True)
      
    print('F1-Score: ', round(f1_score(y_test, pred)*100,2))
```

```
# Defining the metrics
def metrics1(model_name, model, X_test, y_test, pred):
    
    metrics(model_name, y_test, pred)

    lc = model.predict_proba(X_test)
    lc = lc[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, lc)
    plt.plot(fpr, tpr, marker='.')
    plt.title('ROC Curve')
    plt.show()

    print('AUC for ROC: ', roc_auc_score(y_test, lc))
    
    precision, recall, thresholds = precision_recall_curve(y_test, lc)
    plt.plot(recall,precision, marker='.')
    plt.title('Precision-Recall Curve')
    plt.show()
    
    print('AUC for Precision & Recall: ', auc(recall,precision))
```

#### ___Dimenstionality Reduction___

##### ___Principal Component Analysis___

```
#  Principal Component Analysis
from sklearn.decomposition import PCA

target_pca = cancer_df['target']
data_pca = cancer_df.drop('target', axis=1)

target_pca = pd.DataFrame(target_pca)

#To make a PCA, normalize data is essential
X_pca = data_pca.values
X_std = StandardScaler().fit_transform(X_pca)

pca = PCA(svd_solver='full') # https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
pca_std = pca.fit(X_std, target_pca).transform(X_std)

pca_std = pd.DataFrame(pca_std)

# merging the pca values and target
pca_std = pca_std.merge(target_pca, left_index = True, right_index = True, how = 'left')
pca_std['target'] = pca_std['target'].replace({1:'malignant',0:'benign'})
```

```
pca_std.head()
```

```
#explained_variance 
var_pca = pd.DataFrame(pca.explained_variance_ratio_)
var_pca = var_pca.T
round(var_pca*100,2)
```

_After performing PCA, first 5 to 6 columns explains good amount of variance._

###### ___Selecting PCA Components___

```
variance_list = [0.95,0.90,0.85,0.80,0.75,0.70]
```

```
for variance in variance_list:
    pca = PCA(variance)
    pca.fit(X_std)
    n_components = pca.n_components_
    print('Opted Variance: {}, PCA Components: {}'.format(variance,n_components))
```

_Using PCA components in Model. Refer: https://acadgild.com/blog/logistic-regression-with-pca-algorithm_
